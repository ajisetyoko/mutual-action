```python
import numpy as np
```

```python
data = np.loadtxt('res.csv',delimiter=',')
```

```python
data
```

```python
target = data[0,:].astype(int)
```

```python
result = data[1,:].astype(int)
```

```python
result
```

```python
corect = 0
wrong = 0
for i in range(target.shape[0]):
#     print(target[i])
    if target[i] != 51:
        continue
    elif target[i] ==51:
#         print(result[i])
        if target[i]==result[i]:
            corect += 1
        else:
            wrong  += 1
```

```python
corect
```

```python
wrong
```

```python
import pandas as pd
y_actu = pd.Series(target, name='Actual')
y_pred = pd.Series(result, name='Predicted')
df_confusion = pd.crosstab(y_actu, y_pred)
```

```python
df_confusion
```

```python
actual = df_confusion.index.tolist()
for i,class_name in enumerate(actual):
    pred_true_side = df_confusion[class_name]
    true_pos  = pred_true_side.values[i]
    precision = true_pos/(pred_true_side.values).sum()
    actual_true_side = df_confusion.loc[class_name]
    recall = true_pos/(actual_true_side.values).sum()
    print('Class Name {:3d} Precission {:.2f} Recall {:.2f}'.format(class_name,precision,recall))
```
